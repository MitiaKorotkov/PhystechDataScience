## Задача 1.

Пусть $X_1,...,X_n$ выборка из некоторого распределения $\mathrm{P}$, причем ${\sf D} X_1=\sigma^2<+\infty$, и $\sigma$ неизвестно.
Рассмотрим статистику $S^2=\frac1n\sum\limits_{i=1}^n(X_i-\overline{X})^2$.

1. Докажите, что статистика $S^2$ равна $\overline{X^2}-\overline{X}^2$.

$$S^2 = \frac{1}{n}\sum_{i = 1}^{n}(X_i - \overline{X})^2 = \frac{1}{n}\sum_{i = 1}^{n}(X_i^2 - 2X_i\overline{X} + \overline{X}^2) = \frac{1}{n}\sum_{i = 1}^{n}X_i^2 - 2\frac{1}{n}\sum_{i = 1}^{n}X_i\overline{X} + \frac{1}{n}\sum_{i = 1}^{n}\overline{X}^2 = \\ = \overline{X^2} - 2\overline{X}\frac{1}{n}\sum_{i = 1}^{n}X_i + \frac{1}{n}\sum_{i = 1}^{n}\overline{X}^2 = \overline{X^2} - 2\overline{X}^2 + \overline{X}^2 = \overline{X^2} - \overline{X}^2$$

2. Является ли она состоятельной оценкой $\sigma^2$? Для ответа на этот вопрос воспользуйтесь утверждениями с лекции.

Да, она состоятельная. Согласно ЗБЧ, $\overline{X^2_n}\xrightarrow{P}\mathbb{E}\xi^2$ и $\overline{X}_n\xrightarrow{P}\mathbb{E}\xi$. Докажем, что если $\xi_n\xrightarrow{P}\xi$ и $\chi_n\xrightarrow{P}\chi$, тогда и $\xi_n + \chi_n \xrightarrow{P} \xi + \chi$. А также, что если $\xi_n\xrightarrow{P}\xi$ то и $\chi_n\xrightarrow{P}\chi$. Тогда, вспомнив, что $S^2_n = \overline{X_n^2} - \overline{X_n}^2$, немедленно получим состоятельность исследуемой оценки: $\overline{X_n}^2\xrightarrow{P}(\mathbb{E}\xi)^2,~\overline{S_n} = \overline{X_n^2} - \overline{X_n}^2 \xrightarrow{P}\mathbb{E}\xi^2 - (\mathbb{E}\xi)^2 = \sigma^2$.

1. $\xi_n\xrightarrow{P}\xi,~\chi_n\xrightarrow{P}\chi\quad\Rightarrow\quad\xi_n + \chi_n \xrightarrow{P} \xi + \chi$
    
    Пусть $\varepsilon > 0$.
    $$\mathbb{P}(|\xi_n + \chi_n - \xi - \chi| > \varepsilon) \leq \mathbb{P}(|\xi_n - \xi| + |\chi_n - \chi| > \varepsilon)\leq \mathbb{P}(|\xi_n - \xi| > \frac{\varepsilon}{2}) + \mathbb{P}(|\chi_n - \chi| > \frac{\varepsilon}{2})\xrightarrow{P}0\qquad\Box$$

2.  $\xi_n\xrightarrow{P}\xi\qquad\Rightarrow\qquad\chi_n\xrightarrow{P}\chi$

    Зафиксируем $\varepsilon > 0$, такое что $|\xi_n - \xi| \leq \frac{\varepsilon}{|\xi|}$.

    $$|\xi_n^2 - \xi^2| = |\xi_n - \xi|\cdot|\xi_n + \xi| = |\xi_n - \xi|\cdot|\xi_n - \xi + 2\xi| \leq |\xi_n - \xi|\cdot\left(|\xi_n - \xi| + 2|\xi|\right) \leq \\[5pt] \leq \frac{\varepsilon}{|\xi|}\left(\frac{\varepsilon}{|\xi|} + 2|\xi|\right)$$

    Для произвольного $\delta > 0$ оценим $|\xi| \leq K$ так, тобы $M > 1$ и $\mathbb{P}(|\xi| > M) < \delta$. Тогда 
    $$\frac{\varepsilon}{|\xi|}\left(\frac{\varepsilon}{|\xi|} + 2|\xi|\right) \leq \frac{\varepsilon}{M}\left(\frac{\varepsilon}{M} + 2M\right) \leq \frac{\varepsilon}{M}\left(1 + 2M\right) \leq 3\varepsilon$$

    $$\mathbb{P}(|\xi_n^2 - \xi^2| > 3\varepsilon) \leq \mathbb{P}(|\xi_n - \xi| > \varepsilon) + \mathbb{P}(|\xi| > M)$$

    Первое слагаемое тут можно сделать меньше $\delta$ в силу определения сходимости по вероятности, поэтому всю сумму можно оценить $2\delta$

    В итоге получаем требуемое

    $$\mathbb{P}(|\xi_n^2 - \xi^2| > 3\varepsilon) < 2\delta$$

3. Является ли статистика $S^2$ несмещенной оценкой $\sigma^2$? Для ответа на этот вопрос вспомните различные варианты определения дисперсии.

Воспользуемся линейностью матожидания

$$\mathbb{E}S^2 = \mathbb{E}\frac{1}{n}\sum_{i = 1}^{n}\left(\xi_i - \frac1n\sum_{i = 1}^{n}\xi_i\right)^2 = \mathbb{E}\frac{1}{n}\sum_{i = 1}^{n}\left(\xi_i^2 - \frac{2\xi_i}{n}\sum_{j=1}^{n}\xi_j + \frac{1}{n^2}\left(\xi_1^2 + \dots + \xi_n^2 + \sum_{i\neq j}\xi_i\xi_j\right)\right) = \\ = \frac{1}{n}\sum_{i = 1}^{n}\left(\mathbb{E}\xi_i^2 - \frac{2}{n}\mathbb{E}\left(\xi_i^2 + \xi_i(\xi_1 + \dots + \hat{\xi}_j + \dots + \xi_n)\right) + \frac{1}{n^2}\left(\mathbb{E}\xi_1^2 + \dots + \mathbb{E}\xi_n^2 + \sum_{i\neq j}\mathbb{E}\xi_i\mathbb{E}\xi_j\right)\right) = \\ = \frac{1}{n}\sum_{i = 1}^{n}\left(\mathbb{E}\xi_i^2 - \frac{2}{n}\left(\mathbb{E}\xi_i^2 + \sum_{j \neq i}\mathbb{E}\xi_i\mathbb{E}\xi_j\right) + \frac{1}{n^2}\left(n\mathbb{E}\xi_1^2 + n(n - 1)\mathbb{E}\xi_1\mathbb{E}\xi_1\right)\right)$$

Вспомним, что 

$$\mathbb{D} = \mathbb{E}\xi^2 - (\mathbb{E}\xi)^2 \Rightarrow \mathbb{E}\xi_i^2 = \sigma^2 + (\mathbb{E}\xi_i)^2$$

Раскрывая скобки и подставляя, а также пользуясь тем, что матожидание и вторые моменты у всех случайных величин одинаковые получаем

$$\frac{1}{n}\sum_{i = 1}^{n}\mathbb{E}\xi_1^2 - \frac{2}{n^2}\sum_{i = 1}^{n}\mathbb{E}\xi_i^2 - \frac{2(n - 1)}{n^2}\sum_{i = 1}^{n}(\mathbb{E}\xi_1)^2 + \frac{1}{n^2}\sum_{i = 1}^{n}\mathbb{E}\xi_1^2 + \frac{n - 1}{n^2}\sum_{i = 1}^{n}(\mathbb{E}\xi_1)^2 = \\[5pt] = \mathbb{E}\xi_1^2 - \frac{2}{n}\mathbb{E}\xi_1^2 - \frac{2(n - 1)}{n}(\mathbb{E}\xi_1)^2 + \frac{1}{n}\mathbb{E}\xi_1^2 + \frac{n - 1}{n}(\mathbb{E}\xi_1)^2 = \frac{n - 1}{n}\mathbb{E}\xi_1^2 - \frac{n - 1}{n}\mathbb{E}(\xi_1)^2 = \frac{n - 1}{n}\sigma^2$$

Значит эта оценка смещённая

  $$D A \psi = A (D \psi) A^T$$

$$\mathbb{D}\widehat{\theta} = \mathbb{D}(X^TX)^{-1}X^TY = \mathbb{D}(X^TX)^{-1}X^T(X \theta + \varepsilon) = \mathbb{D}(X^TX)^{-1}X^T\varepsilon = (X^TX)^{-1}X^T\left(\mathbb{D}\varepsilon\right)((X^TX)^{-1}X^T)^T = \\[5pt] = (X^TX)^{-1}X^T\left(\sigma^2I\right)X(X^TX)^{-1} = \sigma^2(X^TX)^{-1}X^TX(X^TX)^{-1} = \sigma^2 (X^TX)^{-1}$$
Мультиколлинеарность приведёт к тому, что матрица $X^TX$ будет плохо обусловлена, а погрешность вычисления параметров модели при этом будет квадратично зависеть от числа обусловленности этой матрицы, то есть будет очень велика. Иными словами, чем хуже обусловлена матрица $X$, тем больше будет матрица ковариаций оценок, то есть тем больше будут дисперсии оценок параметров. То есть даже при хороших, но различных выборках мы будем получать сильно отличающиеся друг от друго оценки параметров.

Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$. Известно, что в качестве оценки параметра $\sigma^2$ можно использовать следующие оценки $S^2, \frac{n}{n-1}S^2$. Проверьте эти оценки на несмещенность повторив эксперимент с семинара.

Реализуйте две функции, каждая из которых на вход берет несколько выборок, а на выход выдает массив оценок  
(первая функция - $S^2$, вторая функция - $\frac{n}{n-1}S^2$)  для каждой выборки.  
*Примечание:* У матрицы, заданной в виде `numpy.array` есть метод `var` вычисляющий выборочную дисперсию.

```python
def estimate_s2(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    return X.var(axis=1)

def estimate_s2_corrected(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    n = X.shape[1]
    return n * X.var(axis=1) / (n - 1)
```

```python
def est_plot(theta, true_theta, sample_size):
    '''
    Строит графики как в задаче 5 для каждой из координат тета
    '''
    plt.figure(figsize=(15, 0.7 * len(theta)))
    
    for i, theta_i in enumerate(theta):
        theta_i = theta_i[:sample_size]
        color = np.random.choice(['green', 'blue', 'purple', 'yellow', 'cyan', 'violet', 'orange'])
        plt.scatter(theta_i, np.zeros(sample_size) + i, alpha=0.1, s=100, color=color, label=f'$\theta_{i}$')
        plt.scatter(theta_i.mean(), i, marker='*', s=300, color='w', edgecolors='black')
        plt.vlines(true_theta[i], 0, 1.1 * len(true_theta), color='r')
    
    plt.legend()
    plt.yticks([])
    plt.title('Размер выборки = %d' % sample_size)
    plt.xlim((min(true_theta) - 0.2, max(true_theta) + 0.2))
    plt.ylim((-0.5, 1.1 * len(true_theta)))
```

```python
def generate_X(theta, n_objects, limits=(-100, 100)):
    '''
    Возвращает матрицу плана для линейной регрессии, заполненную случайными числами
    '''
    return np.random.randint(*limits, size=(n_objects, len(theta)))
```

```python
def generate_multicollinear_X(theta, n_objects, limits=(-100, 100)):
    '''
    Возвращает плохо обусловленную матрицу плана для линейной регрессии
    '''
    return np.array([[2, 1], [2.001, 1]])
```

```python
def generate_Y(
    X,
    true_thetas,
    noise_distribution,
    n_objects,
    n_samples,
    distribution_scale=1,
):
    '''
    Возвращает зашумлённый вектор таргетов для линейной регрессии
    '''
    theta = np.array(true_thetas)
    epsilon = noise_distribution.rvs(
        size=(n_samples, n_objects), scale=distribution_scale
    )

    return X @ theta + epsilon
```

Сгенерируем Данные для линейной регрессии и сделаем предсказания с помощью обычной и ридж регрессий:
$$\hat{\theta}_{MSE} = (X^TX)^{-1}X^TY$$
$$\hat{\theta}_{Ridge} = (X^TX + \alpha E)^{-1}X^TY$$

```python
alpha = 1000
theta = [1, 1]
n_objects = 10
X = generate_X(theta, n_objects)
Y = generate_Y(X, theta, sps.norm, n_objects, 1000, distribution_scale=5)

mse_theta = np.linalg.inv(X.T @ X) @ X.T @ Y.T
ridge_theta = np.linalg.inv(X.T @ X + alpha * np.eye(len(theta))) @ X.T @ Y.T
```

Так как параметр двумерный можно нарисовать его предсказания на диаграмме рассеяния. Сделаем это для обеих регрессий

```python
plt.scatter(*mse_theta)

plt.scatter(*mse_theta.mean(axis=1), marker='*', s=300, color='w', edgecolors='black')
plt.scatter(*theta, marker='o', s=3, color='r')
```

```python
plt.scatter(*ridge_theta)

plt.scatter(*ridge_theta.mean(axis=1), marker='*', s=300, color='w', edgecolors='black')
plt.scatter(*theta, marker='o', s=3, color='b')
plt.show()
```

Видим, что при большом коэффициенте регуляризации ридж регрессия не сходится к истинному значению параметра, а обычная -- сходится. Это и есть несмещённость. Получается, что практические результаты вновь совпали с теоретическими
