* С помощью `np.all` проверьте, что в ответе только единицы.

```python
N = 100

nums = np.random.uniform(-2 * np.pi, 2 * np.pi, N)
trig_eq = np.sin(nums) ** 2 + np.cos(nums) ** 2

(
    print("Математика работает!")
    if not np.all(trig_eq - 1)
    else print("Математика сломалась")
)
```

2. Cоздайте матрицы $\begin{pmatrix} -3 & 4 \\ 4 & 3 \end{pmatrix}$ и $\begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix}$. Посчитайте их поэлементное и матричное произведения.

```python
A = np.array([[-3, 4], [4, 3]])
B = np.array([[2, 1], [1, 2]])

el_wise_prod = np.multiply(A, B)
matrix_prod = A @ B

print(f"Поэлементное произведение: \n{el_wise_prod}")
print(f"Матричное произведение: \n{matrix_prod}")
```

3. В статистике и машинном обучении часто приходится иметь с функцией $RSS$, которая вычисляется по формуле $\sum_{i=1}^{n} (y_i - a_i)^2$, где $y_i$ — координаты одномерного вектора $y$,  $a_i$ — координаты одномерного вектора $a$. Посчитайте $RSS$ для $y = (1, 2, 3, 4, 5), a = (3, 2, 1, 0, -1)$.

```python
def rss(y: np.ndarray, a: np.ndarray) -> float:
    return np.sum((y - a) ** 2)


y = np.array([1, 2, 3, 4, 5])
a = np.array([3, 2, 1, 0, -1])

print(rss(y, a))
```

## Задача 3 (<a  target="_blank" href="http://docs.scipy.org/doc/scipy/reference/stats.html">scipy.stats</a>):

${U}(a, b)$ — равномерное распределение. 
  
Параметры в `scipy.stats`:  
* `loc` = $a$,
* `scale` = $b-a$.

Свойства распределения: 
* математическое ожидание: $\frac{a+b}{2}$,
* дисперсия: $\frac{(b-a)^2}{12}$.

1. Постройте график плотности равномерного распределения ${U}(0, 1)$:

```python
plt.figure(figsize=(15, 5))

grid = np.linspace(-2, 2, 500)

plt.plot(grid, sps.uniform.pdf(grid), color='purple', label='Плотность равномерного распределения')
plt.legend()

plt.show()
```

2. Сгенерируйте данные из равномерного распределения ${U}(0, 1)$ и сравните гистограмму с плотностью :

Построим гистограммы для выборок размерами от 100 до 1000000 элементов

```python
sizes = np.logspace(start=2, stop=6, num=5, base=10, dtype=int)
samples = [sps.uniform.rvs(size=s) for s in sizes]
grid = np.linspace(-2, 2, 500)

fig, axs = plt.subplots(5, 1, figsize=(15, 20))

for i in range(len(axs)):
    axs[i].hist(samples[i], bins=100, density=True)
    axs[i].set_title(f"sample {i + 1}: size = {sizes[i]}")
    axs[i].plot(grid, sps.uniform.pdf(grid), color='purple', label='Истинная плотность')
    
fig.suptitle("ОООО, мои гистограммы... \n Нормированная гистограмма и график истинной плотности")
plt.legend()
plt.show()
```

3. Постройте плотности распределений ${U}(0, 1)$, ${U}(0, 2)$, ${U}(4, 1)$ на одном графике, опишите как плотность распределения зависит от его параметров:

```python
params = [(0, 1), (0, 2), (4, 1)]
grid = np.linspace(-1, 6, 500)
colors = ["green", "purple", "blue"]

plt.figure(figsize=(15, 5))

for i, p in enumerate(params):    
    plt.plot(grid, sps.uniform(*p).pdf(grid), color=colors[i], label=f'a = {p[0]}, b = {p[1]}')

plt.title("Графики плотностей равномерных распределений с различными параметрами")
plt.legend()

plt.show()
```

**Ответ:** График плотности распределения представляет из себя отмасштабированную функцию Хэвисайда. Причём такую, что площадь под ступенькой всегда равна 1. При этом, при $a < b$ левый край ступеньки находится в точке с координатой $a$, а правый $-$ в точке с координатой $b$. При $a = b$ график представляет из себя дельта функцию, что в рамках компьютерной арифметики вызыват zero division error. Если $a > b$, то координата левого конца ступеньки $-$ $b$, а правого $-$ $a + b$, поэтому при $b = 0$ мы вновь получаем ошибку деления на ноль.

Всё вышесказанное справедливо при положительныых $a$ и $b$. Для отрицательных $b$ график не существует вовсе. При $a < 0$, $b > 0$ мы попадаем в случай $a < b$

Обобщая эти наблюдения можно предположить, что $uniform(a, b)$ соответствует растянутому в $b$  раз и сдвинутому на $a$ вправо стандартному равномерному распределению $[0, 1]$
